Distributions to explore: normal, binomial, chi-square, exponential, lognormal, students-t (standard_t)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#normal distribution parameters are optional 
normalvar = np.random.normal(50, 300, 100)
print('mean: ' + str(normalvar.mean()))
print('std deviation: ' + str(normalvar.std()))

plt.hist(normalvar)

#axvline draws a vertical line, general params are x, ymin and ymax
plt.axvline(normalvar.mean(), color='c', linestyle='solid', linewidth=2)
plt.axvline(normalvar.mean() + normalvar.std(), color='c', linestyle='dashed', linewidth=2)
plt.axvline(normalvar.mean() - normalvar.std(), color='c', linestyle='dashed', linewidth=2)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#binomial parameters are (n, p)
binomialvar = np.random.binomial(5, .5, 100)
print('mean: ' + str(binomialvar.mean()))
print('std deviation: ' + str(binomialvar.std()))

plt.hist(binomialvar)

#axvline draws a vertical line, general params are x, ymin and ymax
plt.axvline(binomialvar.mean(), color='c', linestyle='solid', linewidth=2)
plt.axvline(binomialvar.mean() + binomialvar.std(), color='c', linestyle='dashed', linewidth=2)
plt.axvline(binomialvar.mean() - binomialvar.std(), color='c', linestyle='dashed', linewidth=2)
plt.show()
```

```python
#chisquare parameter is degrees freedom [and size]. Degrees freedom is num classes - 1
chisvar = np.random.chisquare(3, 100)
print('mean: ' + str(chisvar.mean()))
print('std deviation: ' + str(chisvar.std()))

plt.hist(chisvar)

#axvline draws a vertical line, general params are x, ymin and ymax
plt.axvline(chisvar.mean(), color='c', linestyle='solid', linewidth=2)
plt.axvline(chisvar.mean() + chisvar.std(), color='c', linestyle='dashed', linewidth=2)
plt.axvline(chisvar.mean() - chisvar.std(), color='c', linestyle='dashed', linewidth=2)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#exponential parameter is scale -- scale is 1- lambda [and size]. 
expvar = np.random.exponential(0.98, 100)
print('mean: ' + str(expvar.mean()))
print('std deviation: ' + str(expvar.std()))

plt.hist(expvar)

#axvline draws a vertical line, general params are x, ymin and ymax
plt.axvline(expvar.mean(), color='c', linestyle='solid', linewidth=2)
plt.axvline(expvar.mean() + expvar.std(), color='c', linestyle='dashed', linewidth=2)
plt.axvline(expvar.mean() - expvar.std(), color='c', linestyle='dashed', linewidth=2)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#log-normal parameters are mean, sigma, size
lnvar = np.random.lognormal(60.01, 1,100)
print('mean: ' + str(lnvar.mean()))
print('std deviation: ' + str(lnvar.std()))

plt.hist(lnvar)

plt.axvline(lnvar.mean(), color='c', linestyle='solid', linewidth=2)
plt.axvline(lnvar.mean() + lnvar.std(), color='c', linestyle='dashed', linewidth=2)
plt.axvline(lnvar.mean() - lnvar.std(), color='c', linestyle='dashed', linewidth=2)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#students t parameters are degrees freedom, size
stvar = np.random.standard_t(3, 100)
print('mean: ' + str(stvar.mean()))
print('std deviation: ' + str(stvar.std()))

plt.hist(stvar)

plt.axvline(stvar.mean(), color='c', linestyle='solid', linewidth=2)
plt.axvline(stvar.mean() + stvar.std(), color='c', linestyle='dashed', linewidth=2)
plt.axvline(stvar.mean() - stvar.std(), color='c', linestyle='dashed', linewidth=2)
plt.show()
```

Evaluate whether the descriptive statistics (mean, standard deviation) provided useful information about the variable. Yes ==> normal, student's t. No ==> Binomial, at least when sample size is small relative to population, chi-square, exponential, log normal. Logarithmic and exponential distributions are skewed so we should not expect descriptive statistics to accurately describe them.
